```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
water_Tank=1000
df=pd.DataFrame({
    'waterconsumption':[110,120,150,160,170,180,190],#litres
    'days':['mon','tue','wed','thu','fri','sat','sun']
})
print("------Dataset-------")
print(df)
daily_increment=[]
for i in range(len(df['waterconsumption'])-1):
  daily_increment.append(df['waterconsumption'][i+1]-df['waterconsumption'][i])
print(f"Daily water consumption: {daily_increment}")
avg=sum(daily_increment)/len(daily_increment)
print(f"Average water consumption: {avg}")
if(df["waterconsumption"][len(df["waterconsumption"])-1]+avg)>water_Tank:
  print("Not Sufficient")
else:
  print("We can use water")
  # Histogram
  plt.plot(df['days'],df['waterconsumption'])
  plt.show()
  # Bar Plot
  plt.bar(df['days'],df['waterconsumption'])
  plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the weather history dataset
df_weather = pd.read_csv('/content/weatherHistory.csv')

df_weather['Formatted Date'] = pd.to_datetime(df_weather['Formatted Date'])

df_weather = df_weather.set_index('Formatted Date')

print("------Weather Data Overview-------")
print(df_weather.head())
print("\n------DataFrame Information-------")
df_weather.info()
```

```python
plt.figure(figsize=(10, 6))
df_weather_subset = df_weather.loc[df_weather.index.min():df_weather.index.min() + pd.Timedelta(days=7)]

plt.plot(df_weather_subset.index, df_weather_subset['Temperature (C)'])
plt.xlabel("Date and Time")
plt.ylabel("Temperature (C)")
plt.title("Hourly Temperature Trend (First 7 Days)")
plt.grid(True)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
